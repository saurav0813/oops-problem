Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
import logging
logging.basicConfig(filename="assignment.log",level=logging.DEBUG,format = '%(asctime)s - %(levelname)s - %(message)s')
class SBI:
    def __init__(self,an,ahn,b):
        logging.info("init constructor initialize the attributes")
        self.account_number = an
        self.account_holder_name = ahn
        self.balance = b
        self.bank_name = "State Bank Of India"

    def deposit(self,amount):
        logging.info(f"depositing {amount} to the account {self.account_number}")
        try:
            if amount<=0:
                raise ValueError("amount should be positive.")
            self.balance += amount
            logging.info(f"Deposited amount {amount} to the account xxxxxx{self.account_number[-4::]}. Avl/bal in a/c xxxxxx{self.account_number[-4::]} is {self.balance}")
        except ValueError as v:
            logging.error(f"value error : {v}")
        except Exception as e:
            logging.error(f"error is {e}")
    def withdrawl(self,amount):
        try:
            logging.info("Attempting to withdraw money {amount} from account xxxxxx{self.account_number[-4::]}")
            if amount <= 0:
                raise ValueError("amount must be positive")
            elif amount>self.balance:
                logging.warning(f"you've no sufficient amount...")
            
            else:
                self.balance -= amount
                logging.info(f"deducted amount is {amount}. avl/bal to a/c xxxxxx{self.account_number[-4::]} is {self.balance}")
        except ValueError as v:
            logging.error(f"ValueError : {v}")
        
        except Exception as e:
            logging.error(f"unexpected error : {e}")


customer1 = SBI("12345678","saurav",300000)
customer1.deposit(500000)
customer1.withdrawl(40000)

print(f"Final balance for account xxxxxx{customer1.account_number[-4:]}: {customer1.balance}")

    
```

Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

```python
import logging

logging.basicConfig(filename="vehicle_rental_system.log", level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')

class Vehicle:
    def __init__(self, vehicle_id: int, vehicle_type: str, brand: str, model: str):
        self.vehicle_id = vehicle_id
        self.vehicle_type = vehicle_type
        self.brand = brand
        self.model = model
        self.rented = False

    def __str__(self):
        return f"ID: {self.vehicle_id}, Type: {self.vehicle_type}, Brand: {self.brand}, Model: {self.model}, Rented: {self.rented}"

class VehicleRentalSystem:
    def __init__(self):
        logging.info("Initializing an empty list to add vehicles")
        self.vehicles = []

    def add_vehicle(self, vehicle: Vehicle):
        logging.info("Checking if vehicle is an instance of Vehicle class")
        try:
            if not isinstance(vehicle, Vehicle):
                raise ValueError("The provided vehicle is not an instance of Vehicle class")
        except ValueError as v:
            logging.error(f"ValueError: {v}")
        else:
            self.vehicles.append(vehicle)
            logging.info(f"{vehicle} added successfully.")

    def rent_vehicle(self, vehicle_id: int) -> str:
        logging.info(f"Attempting to rent a vehicle with ID {vehicle_id}")
        try:
            for vehicle in self.vehicles:
                if vehicle.vehicle_id == vehicle_id:
                    if not vehicle.rented:
                        vehicle.rented = True
                        logging.info(f"Vehicle with ID {vehicle_id} rented successfully.")
                        return f"Vehicle {vehicle_id} rented successfully."
                    else:
                        logging.warning(f"Vehicle with ID {vehicle_id} is already rented.")
                        return f"Vehicle {vehicle_id} is already rented."
            logging.warning(f"Vehicle with ID {vehicle_id} not found.")
            return f"Vehicle {vehicle_id} not found."
        except Exception as e:
            logging.error(f"Unexpected error: {e}")
            return "An unexpected error occurred."

    def return_vehicle(self, vehicle_id: int) -> None:
        logging.info(f"Attempting to return vehicle with ID {vehicle_id}")
        for vehicle in self.vehicles:
            if vehicle.vehicle_id == vehicle_id:
                if vehicle.rented:
                    vehicle.rented = False
                    logging.info(f"Vehicle with ID {vehicle_id} returned successfully.")
                    print(f"Vehicle {vehicle_id} returned successfully.")
                else:
                    logging.warning(f"Vehicle with ID {vehicle_id} was not rented.")
                    print(f"Vehicle {vehicle_id} was not rented.")
                return
        logging.warning(f"Vehicle with ID {vehicle_id} does not exist.")
        print(f"Vehicle {vehicle_id} does not exist.")

    def display_available_vehicles(self) -> None:
        print("Available Vehicles:")
        available_vehicles = [vehicle for vehicle in self.vehicles if not vehicle.rented]
        if available_vehicles:
            for vehicle in available_vehicles:
                print(vehicle)
        else:
            print("No vehicles available.")

# Example usage:
rental_system = VehicleRentalSystem()

# Add vehicles
rental_system.add_vehicle(Vehicle(1, "Car", "Toyota", "Camry"))
rental_system.add_vehicle(Vehicle(2, "Truck", "Ford", "F-150"))
rental_system.add_vehicle(Vehicle(3, "Motorcycle", "Honda", "CBR500R"))

# Display available vehicles
rental_system.display_available_vehicles()

# Rent vehicle
print(rental_system.rent_vehicle(1))

# Display available vehicles
rental_system.display_available_vehicles()

# Return vehicle
rental_system.return_vehicle(1)
```

Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
import logging

logging.basicConfig(filename="library_catalog.log", level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')

class Book:
    def __init__(self, title: str, author: str, pages: int, book_id: int):
        self.title = title
        self.author = author
        self.pages = pages
        self.book_id = book_id
        self.borrowed = False

    def __str__(self):
        return f"ID: {self.book_id}, Title: {self.title}, Author: {self.author}, Pages: {self.pages}, Borrowed: {self.borrowed}"

class Library:
    def __init__(self):
        logging.info("Initializing an empty list to add books")
        self.books = []

    def add_book(self, book: Book):
        logging.info("Checking if the book is an instance of the Book class")
        try:
            if not isinstance(book, Book):
                raise ValueError("The provided book is not an instance of the Book class")
            self.books.append(book)
            logging.info(f"Book '{book.title}' added successfully.")
        except ValueError as v:
            logging.error(f"ValueError: {v}")

    def borrow_book(self, book_id: int) -> str:
        logging.info(f"Attempting to borrow book with ID {book_id}")
        try:
            for book in self.books:
                if book.book_id == book_id:
                    if not book.borrowed:
                        book.borrowed = True
                        logging.info(f"Book '{book.title}' borrowed successfully.")
                        return f"Book with ID {book_id} borrowed successfully."
                    else:
                        logging.warning(f"Book with ID {book_id} is already borrowed.")
                        return f"Book with ID {book_id} is already borrowed."
            logging.warning(f"Book with ID {book_id} not found.")
            return f"Book with ID {book_id} not found."
        except Exception as e:
            logging.error(f"Unexpected error: {e}")
            return "An unexpected error occurred."

    def display_available_books(self) -> None:
        logging.info("Displaying available books")
        available_books = [book for book in self.books if not book.borrowed]
        if available_books:
            print("Available Books:")
            for book in available_books:
                print(book)
        else:
            print("No books available.")

# Example usage:
library = Library()

# Add books
library.add_book(Book("1984", "George Orwell", 328, 1))
library.add_book(Book("To Kill a Mockingbird", "Harper Lee", 281, 2))
library.add_book(Book("The Great Gatsby", "F. Scott Fitzgerald", 180, 3))

# Display available books
library.display_available_books()

# Borrow a book
print(library.borrow_book(1))

# Display available books again
library.display_available_books()

# Try to borrow the same book again
print(library.borrow_book(1))

# Display available books
library.display_available_books()
```

Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products

```python
import logging

logging.basicConfig(filename="inventory.log", level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')

class Product:
    def __init__(self, pro_id: int, pro_name: str, pro_brand: str, price: float, quantity: int):
        self.product_id = pro_id
        self.product_name = pro_name
        self.product_brand = pro_brand
        self.price = price
        self.quantity = quantity

    def __str__(self) -> str:
        return (f"ID: {self.product_id}, Name: {self.product_name}, Brand: {self.product_brand}, "
                f"Price: {self.price}, Quantity: {self.quantity}")

class Inventory:
    def __init__(self):
        self.products = []
        logging.info("Initializing an empty list to add products")

    def add_products(self, product: Product) -> str:
        try:
            logging.info(f"Checking whether {product} is an instance of class Product")
            if not isinstance(product, Product):
                raise ValueError("Provided product is not of Product class")
        except ValueError as v:
            logging.error(f"ValueError: {v}")
        else:
            self.products.append(product)
            logging.info(f"{product} added successfully to inventory.")
        return f"Inventory contains the following products: {[str(p) for p in self.products]}"

    def update_product_quantity(self, product_id: int, new_quantity: int) -> str:
        logging.info(f"Attempting to update the quantity of the product with ID {product_id}")
        try:
            if not isinstance(new_quantity, int):
                raise ValueError("Quantity must be an integer")
            if new_quantity < 0:
                raise ValueError("Quantity must be greater than or equal to zero")

            for product in self.products:
                if product.product_id == product_id:
                    product.quantity = new_quantity
                    logging.info(f"Product with ID {product_id} updated successfully. New quantity: {product.quantity}")
                    return f"Product ID {product_id} updated successfully. New quantity: {product.quantity}"
            
            logging.warning(f"Product with ID {product_id} not found in inventory")
            return f"Product with ID {product_id} not found in inventory"
        
        except ValueError as ve:
            logging.error(f"ValueError: {ve}")
            return f"Error: {ve}"
        except Exception as e:
            logging.error(f"Unexpected error: {e}")
            return "An unexpected error occurred."
        
    def display_products(self) -> str:
        logging.info("Displaying all products in the inventory")
        if not self.products:
            logging.info("No products available in the inventory.")
            return "No products available in the inventory."

        result = "Product List:\n"
        for product in self.products:
            result += str(product) + "\n"
        logging.info("Displayed all products successfully.")
        return result

# Example usage
if __name__ == "__main__":
    # Create products
    product1 = Product(1, "Oil", "Gemini", 123.00, 20)
    product2 = Product(2, "Biscuit", "Oreo", 10.00, 30)

    # Create inventory and add products
    oil_stack = Inventory()
    print(oil_stack.add_products(product1))
    print(oil_stack.add_products(product2))

    # Display products
    print(oil_stack.display_products())

    # Update product quantity
    print(oil_stack.update_product_quantity(1, 25))  # Valid update
    print(oil_stack.update_product_quantity(2, -5))  # Invalid quantity
    print(oil_stack.update_product_quantity(3, 10))  # Product not found

    # Display products again to see updates
    print(oil_stack.display_products())


```

Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
import math
class Circle:
    def __init__(self,r):
        self.radius = r

    def area(self):
        try:
            logging.info("try to find the area of circle")
            if not isinstance(self.radius,(float,int)):
                raise ValueError(f"{self.radius} should be of int type or float type ")
            area = math.pi*(self.radius**2)
            logging.info(f"finding area of circle {area:.2f}")
            return area
        except ValueError as ve:
            logging.error(f"value error : {ve}")
            return f"radius should be of int or float data type"
    def perimeter(self):
        try:
            logging.info("Attempting to find the perimeter(circumference)")
            if not isinstance(self.radius,(float,int)):
                raise ValueError(f"{self.radius} should be of int type or float type ")
            perimeter = 2*(math.pi)*(self.radius)
            logging.info(f"finding perimeter of circle {perimeter:.2f}")
            return perimeter
        except ValueError as ve:
            logging.error(f"value error : {ve}")
            return f"radius should be of int or float data type"
    
    def semi_perimeter(self):
        try:
            logging.info("Attempting to find the semi perimeter")
            if not isinstance(self.radius,(float,int)):
                raise ValueError(f"{self.radius} should be int type or float type ")
            semi_perimeter = (math.pi)*(self.radius)+2*(self.radius)
            logging.info(f"finding semi perimeter of circle {semi_perimeter:.2f}")
            return semi_perimeter
        except ValueError as ve:
            logging.error(f"value error : {ve}")
            return f"radius should be of int or float data type"

class Rectangle:
    def __init__(self,l,b):
        self.length = l
        self.breadth = b

    def area(self):
        logging.info("checking whether length and breadth are of int or float type")
        try:
            if not (isinstance(self.length,(int,float)) and isinstance(self.breadth,(int,float))):
                raise ValueError("Provided length and breadth are not type of int or float type.")
            logging.info("Attempting to find the area of rectangle")
            area = self.length*self.breadth
            logging.info(f"area of rectangle is {area:.2f}")
            return area
        except ValueError as ve:
            logging.error(f"Value Error : {ve}")
            return f"Value error : {ve}"

    def perimeter(self):
        logging.info("Attempting to find the perimeter of rectangle")
        try:
            logging.info("checking whether length and breadth are of int or float type")
            if not (isinstance(self.length,(int,float)) and isinstance(self.breadth,(int,float))):
                raise ValueError("Provided length and breadth are not type of int or float type.")
            logging.info("Attempting to find the perimeter of rectangle")
            perimeter = 2*(self.length+self.breadth)
            logging.info(f"perimeter of rectangle is {perimeter:.2f}")
            return perimeter
        except ValueError as ve:
            logging.error(f"Value Error : {ve}")
            return f"Value error : {ve}"

class Square:
    def __init__(self,a):
        self.side = a


    def area(self):
        logging.info("Attempting to find the area of square")
        try:
            logging.info("checking whether side is of int or float data type.")
            if not isinstance(self.side,int):
                raise ValueError("povided side is not of int or flot data type")
        except ValueError as ve:
            logging.error(f"value error : {ve}")
            return None
        else:
            area = self.side**2
            logging.info(f"area of square is {area:.2f}")
            return area

    def perimeter(self):
        logging.info("Attempting to find the perimeter of square")
        try:
            logging.info("checking whether side is of int or float data type.")
            if not isinstance(self.side,int):
                raise ValueError("povided side is not of int or flot data type")
        except ValueError as ve:
            logging.error(f"value error : {ve}")
            return None
        else:
            Perimeter = 4*self.side
            logging.info(f"perimeter of square is {Perimeter:.2f}")
            return Perimeter



rectangle1 = Rectangle(2,4)
rectangle2 = Rectangle((1,),"str")
print(rectangle2.area())
print(rectangle1.area())
print(rectangle1.perimeter())
circle1 = Circle(1)
print(circle1.area())
circle2 = Circle("str")
print(circle2.area())
square1 = Square(4)
print(square1.area())
print(square1.perimeter())
```

Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

```python
class Student:
    def __init__(self,s_id,name):
        self.student_id = s_id
        self.name = name
        self.grades = []
    
    def add_grades(self,grade):
        logging.info("checking whether grade is of int or float data type.")
        try:
            if not isinstance(grade,(int,float)):
                raise ValueError("grade is not of int or float data type")
            elif grade < 0:
                raise ValueError("grade shouldn't be negative")
        except ValueError as ve:
            logging.error(f"value error : {ve}")
            return f"Value error : {ve}"
        except ValueError as v:
            logging.error(f"value error : {v}")
            return f"value error : {v}"
        else:
            self.grades.append(grade)
            logging.info(f"grade is added successfully to {self.grades}")
    
    def calculate_average_grade(self):
        try:
            if not self.grades:
                raise ValueError(f"{grade} not available to calculate average")
                logging.warning(f"grade not found in {self.grades} ")
        except ValueError as ve:
            logging.error(f"value error : {ve}")
            return f"value error : {ve}"
        except Exception as e:
            logging.error(f"unexpected error : {e}")
            return f"unexpected error : {e}"
        else:
            logging.info("Attempting to calculate average  grades")
            sum = 0
            length = len(self.grades)
            for num in self.grades:
                sum += num
                avg = sum/length
            logging.info(f"average grade is {avg}")
            return avg

    def display_student_details(self):
        logging.info("Attempting to display student details.")
        return f"""
        Student ID : {self.student_id}
        Student Name :{self.name}
        Student Grades : {self.grades}"""


student1 = Student("sau123","Saurav")
student1.add_grades(85)
student1.add_grades(90)
print(student1.calculate_average_grade())
student2 = Student("ujj456","Ujjwal")
student2.add_grades(87)
print(student2.calculate_average_grade())

```

Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

```python
import logging

# Configure logging
logging.basicConfig(filename='social_media_profile.log', level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')

class SocialMediaProfile:
    def __init__(self, username: str):
        self.username = username
        self.posts = []  # List to store posts
        logging.info(f"Created profile for username: {self.username}")

    def add_post(self, post_content: str) -> None:
        """Adds a new post to the profile."""
        if post_content:
            self.posts.append(post_content)
            logging.info(f"Added post by {self.username}: {post_content}")
        else:
            logging.error("Attempted to add an empty post")
            raise ValueError("Post content cannot be empty")

    def display_posts(self) -> None:
        """Displays all posts in the profile."""
        logging.info(f"Displaying posts for {self.username}")
        if self.posts:
            print(f"Posts by {self.username}:")
            for idx, post in enumerate(self.posts, start=1):
                print(f"{idx}: {post}")
        else:
            print(f"{self.username} has no posts.")
            logging.info("No posts to display")

    def search_posts(self, keyword: str) -> None:
        """Searches for posts containing a specific keyword."""
        logging.info(f"Searching posts for keyword: '{keyword}' in {self.username}'s profile")
        found_posts = [post for post in self.posts if keyword.lower() in post.lower()]
        if found_posts:
            print(f"Posts containing '{keyword}':")
            for idx, post in enumerate(found_posts, start=1):
                print(f"{idx}: {post}")
        else:
            print(f"No posts found containing '{keyword}'.")
            logging.info(f"No posts found with keyword: {keyword}")

# Example usage:
profile = SocialMediaProfile("john_doe")

# Adding posts
profile.add_post("Hello world! This is my first post.")
profile.add_post("I love programming in Python.")
profile.add_post("Today was a great day at the beach!")

# Displaying all posts
profile.display_posts()

# Searching for posts with a keyword
profile.search_posts("Python")
profile.search_posts("beach")
profile.search_posts("food")  # This keyword does not exist
```

Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
import logging
from datetime import datetime

# Configure logging
logging.basicConfig(filename='todo_list.log', level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')

class Task:
    def __init__(self, description: str, due_date: str):
        self.description = description
        self.due_date = datetime.strptime(due_date, '%Y-%m-%d')
        self.completed = False

    def mark_completed(self):
        self.completed = True
        logging.info(f"Task '{self.description}' marked as completed.")

    def __str__(self):
        status = "✔️" if self.completed else "❌"
        return f"{status} {self.description} (Due: {self.due_date.strftime('%Y-%m-%d')})"

class ToDoList:
    def __init__(self):
        self.tasks = []
        logging.info("Initialized a new ToDo list.")

    def add_task(self, description: str, due_date: str) -> None:
        """Adds a new task to the ToDo list."""
        task = Task(description, due_date)
        self.tasks.append(task)
        logging.info(f"Added task: {task}")

    def mark_task_completed(self, index: int) -> None:
        """Marks a task as completed."""
        try:
            self.tasks[index].mark_completed()
        except IndexError:
            logging.error("Attempted to mark a task as completed that does not exist.")
            print("Error: Task not found.")

    def display_pending_tasks(self) -> None:
        """Displays all pending tasks."""
        logging.info("Displaying pending tasks.")
        pending_tasks = [task for task in self.tasks if not task.completed]
        if pending_tasks:
            print("Pending Tasks:")
            for idx, task in enumerate(pending_tasks):
                print(f"{idx + 1}: {task}")
        else:
            print("No pending tasks.")
            logging.info("No pending tasks to display.")

# Example usage
todo_list = ToDoList()

# Adding tasks
todo_list.add_task("Finish the report", "2024-09-30")
todo_list.add_task("Go grocery shopping", "2024-09-15")
todo_list.add_task("Walk the dog", "2024-09-16")

# Displaying pending tasks
todo_list.display_pending_tasks()

# Marking a task as completed
todo_list.mark_task_completed(1)  # Complete the second task

# Displaying pending tasks again
todo_list.display_pending_tasks()
```
